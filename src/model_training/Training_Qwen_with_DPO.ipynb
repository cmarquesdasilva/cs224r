# Supervised Training

````python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import prepare_model_for_kbit_training, LoraConfig, get_peft_model

model_name = "Qwen/Qwen2.5-7B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype="bfloat16",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True,
)

model.gradient_checkpointing_enable()
model = prepare_model_for_kbit_training(model)

lora_config = LoraConfig(
    r=64,
    lora_alpha=16,
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM",
)
model = get_peft_model(model, lora_config)
````

````python
# Qwen Special tokens
SYSTEM_PROMPT = "You are a helpful assistant."
USER_ROLE = "user"
ASSISTANT_ROLE = "assistant"
SYSTEM_ROLE = "system"
START_TOKEN = "<|im_start|>"
END_TOKEN = "<|im_end|>"
````

````python
# Supervised:
import pandas as pd
from datasets import Dataset

# Load dataset
df = pd.read_csv("dpo_training_set.tsv", sep='\t')

# Rename columns to standard
df = df.rename(columns={
    "base_prompt": "prompt",
    "agent_chosen": "chosen",
    "agent_rejected": "rejected"
})

sft_data = []

for _, row in df.iterrows():
    user_message = row["prompt"].strip()
    chosen_response = row["chosen"].strip()
    rejected_response = row["rejected"].strip()

    prompt_chatml = (
        f"{START_TOKEN}{SYSTEM_ROLE}\n{SYSTEM_PROMPT}{END_TOKEN}\n"
        f"{START_TOKEN}{USER_ROLE}\n{user_message}{END_TOKEN}\n"
        f"{START_TOKEN}{ASSISTANT_ROLE}\n"
    )

    sft_data.append({"text": f"{prompt_chatml}{chosen_response}{END_TOKEN}"})
    sft_data.append({"text": f"{prompt_chatml}{rejected_response}{END_TOKEN}"})

# Convert to HF dataset
sft_dataset = Dataset.from_pandas(pd.DataFrame(sft_data))

def tokenize_fn(examples):
    tokenized = tokenizer(
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=1024,
    )

    pad_id = tokenizer.pad_token_id
    # Build labels and mask padding with -100
    labels = [
        [(tok if tok != pad_id else -100) for tok in seq]
        for seq in tokenized["input_ids"]
    ]

    tokenized["labels"] = labels
    return tokenized

tokenized_sft_dataset = sft_dataset.map(
    tokenize_fn,
    batched=True,
    remove_columns=["text"],
)

tokenized_sft_dataset = sft_dataset.map(tokenize_fn, batched=True)
````

````python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/DRL/project/model/sft/",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=2,
    learning_rate=2e-5,
    num_train_epochs=3,
    logging_steps=10,
    save_steps=100,
    bf16=True,
    fp16=False,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_sft_dataset,
    tokenizer=tokenizer
)

trainer.train()
````

````python
import os
import json
import logging
import pandas as pd
from tqdm import tqdm
import re
import random
import torch

def perturb_temperature(base_temp: float, min_temp: float = 0.0, max_temp: float = 1.2, delta: float = 0.0) -> float:
    """
    Applies a small random perturbation to the base temperature.

    Args:
        base_temp (float): The base temperature to perturb.
        min_temp (float): Minimum allowable temperature.
        max_temp (float): Maximum allowable temperature.
        delta (float): Maximum perturbation up or down.

    Returns:
        float: A new temperature value within [min_temp, max_temp].
    """
    perturb = random.uniform(-delta, delta)
    new_temp = base_temp + perturb
    return max(min_temp, min(max_temp, round(new_temp, 3)))


def convert_to_proper_json_format(raw_response):
    """
    Attempts to extract and parse a JSON object from a model-generated string.

    Args:
        raw_response (str): The raw text output from the model.

    Returns:
        dict: A parsed JSON object.

    Raises:
        ValueError: If parsing fails even after cleaning.
    """
    cleaned_response = re.sub(r"^```(?:json)?|```$", "", raw_response.strip(), flags=re.MULTILINE).strip()

    try:
        return json.loads(cleaned_response)
    except json.JSONDecodeError:
        pass

    cleaned_response = re.sub(r",(\s*[}\]])", r"\1", cleaned_response)

    try:
        return json.loads(cleaned_response)
    except json.JSONDecodeError as e:
        raise ValueError(f"Failed to parse cleaned response as JSON. Error: {e}\nRaw content:\n{cleaned_response}")

def call_local_model(
    prompt: str,
    model,
    tokenizer,
    #temperature: float = 0.0,
    max_tokens: int = 4096,
    top_p: float = 1,
    **generate_kwargs,
) -> str:
    """
    Generate text locally and return ONLY the new tokens (prompt stripped).
    """
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    input_ids = inputs["input_ids"]

    with torch.no_grad():
        output_ids = model.generate(
            **inputs,
            max_new_tokens=max_tokens,
            do_sample=False,
            #temperature=temperature,
            top_p=top_p,
            **generate_kwargs
        )

    # Keep only the completion tokens
    gen_ids = output_ids[0][input_ids.shape[-1] :]
    return tokenizer.decode(gen_ids, skip_special_tokens=True)

def extract_json(response: str) -> dict:
    """
    Extracts the first full JSON object ending after the 'other_facts' key from a given response.
    Assumes that the JSON object starts with '{' and that "other_facts" is the last key.
    """
    start_idx = response.find('{')
    if start_idx == -1:
        raise ValueError("No opening brace found in response.")

    # Find position of "other_facts"
    other_facts_idx = response.find('"other_facts"', start_idx)
    if other_facts_idx == -1:
        raise ValueError('"other_facts" key not found in response.')

    # Start scanning from start_idx and count braces
    brace_count = 0
    for i in range(start_idx, len(response)):
        if response[i] == '{':
            brace_count += 1
        elif response[i] == '}':
            brace_count -= 1
            # If we reached zero after "other_facts", this is the end of our JSON
            if brace_count == 0 and i > other_facts_idx:
                json_str = response[start_idx:i+1]
                try:
                    return json.loads(json_str)
                except json.JSONDecodeError as e:
                    raise ValueError(f"Extracted text is not valid JSON: {e}")

    raise ValueError("Could not find a complete JSON object ending after 'other_facts'.")
````

````python
import pandas as pd
import chevron
from typing import Dict, List
from dataclasses import dataclass, field
import json

@dataclass
class PersonalitySchema:
    traits: List[str] = field(default_factory=lambda: [
        "openness", "conscientiousness", "extraversion", "agreeableness", "neuroticism"
    ])
    facets: Dict[str, List[str]] = field(default_factory=lambda: {
        "openness": [
            "imagination", "artistic_interests", "emotionality",
            "adventurousness", "intellect", "liberalism"
        ],
        "conscientiousness": [
            "self_efficacy", "orderliness", "dutifulness",
            "achievement_striving", "self_discipline", "cautiousness"
        ],
        "extraversion": [
            "friendliness", "gregariousness", "assertiveness",
            "active_level", "excitement_seeking", "cheerfulness"
        ],
        "agreeableness": [
            "trust", "morality", "altruism",
            "cooperation", "modesty", "sympathy"
        ],
        "neuroticism": [
            "anxiety", "anger", "depression",
            "self_consciousness", "immoderation", "vulnerability"
        ]
    })

class PromptBuilder:
    def __init__(self, data: pd.DataFrame, template_path: str, example_path: str = None):
        self.data = data
        self.template_path = template_path
        #self.example_1 = self._load_example(example_path) if example_path else ""
        self.schema = PersonalitySchema()

    def _load_example(self, example_path: str) -> str:
        with open(example_path, 'r') as f:
            example_data = json.load(f)
        return json.dumps(example_data, indent=2)

    def _get_person_data(self, index: int) -> Dict:
        row = self.data.iloc[index]
        # Convert numeric sex to label
        sex = "Female" if row["sex"] == 2 else "Male"

        # Extract trait levels
        traits = {
            trait: row[f"{trait}_level"]
            for trait in self.schema.traits
        }

        # Extract facet levels
        facets = {}
        for trait, facet_list in self.schema.facets.items():
            for facet in facet_list:
                matches = [col for col in self.data.columns if col.startswith(f"{facet}_level")]
                if matches:
                    facets[facet] = row[matches[0]]
                else:
                    facets[facet] = "Unknown"

        return {
            "personality_profile": {
                **traits,
                **facets
            },
            "age": row["age"],
            "sex": sex,
            "country": row["country"]
            #"example_1": self.example_1
        }

    def build_prompt(self, index: int) -> str:
        context = self._get_person_data(index)
        with open(self.template_path, 'r') as f:
            template = f.read()
        return chevron.render(template, context)
````
